```python
import os
import numpy as np
from IPython.display import display
import torch
from data import adult
def statistics(log_path, protected_var):
    n_success = 0
    n_total = 0
    n_query = []
    n_sensitive_change = 0
    with open(os.path.join('log', f'{log_path}.log'),'r') as f:
        lines = f.readlines()
        for line in lines[1:]:
            t, s, n, p = line.split('\t')
            n_success += int(s)
            n_total += 1
            n_query.append(int(n))
            p = eval(p)
            if p != None:
                p = adult.get_original_feature(torch.tensor(p))
                if p[protected_var][0] != p[protected_var][1]:
                    n_sensitive_change += 1
    n_query = np.array(n_query)
    print('success:', f'{n_success}/{n_total}', '\naverage n_query:', n_query.mean(), '\nstd of n_query',  n_query.std(), '\nnum of pairs that changed protected var', n_sensitive_change)
```

```python
statistics('select_RF_range', protected_var='race_White')
```
