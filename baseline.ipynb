```python
import os
import numpy as np
from IPython.display import display
import torch
from data import adult
def statistics(log_path):
    n_success = 0
    n_query = []
    with open(os.path.join('log', f'{log_path}.log'),'r') as f:
        lines = f.readlines()
        for line in lines[1:]:
            t, s, n, p = line.split('\t')
            n_success += int(s)
            n_query.append(int(n))
    n_query = np.array(n_query)
    print('success:', n_success, '\naverage n_query:', n_query.mean(), '\nstd of n_query',  n_query.std())
```

```python
statistics('dx_select')
```
